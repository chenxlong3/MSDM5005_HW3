```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
from webdriver_manager.chrome import ChromeDriverManager
import time
import random
import pandas as pd
```

```python
url = "https://www.midland.com.hk/en/list/buy"
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
```

```python
MAX_PAGE_NUM = driver.find_elements(by=By.XPATH, value="//a[contains(@aria-label, 'Page')]")[-1].text
MAX_PAGE_NUM = int(MAX_PAGE_NUM)
some_timeout = 10
min_len_per_page = 21
ignored_exceptions=(NoSuchElementException,StaleElementReferenceException,)
```

```python

name_list = []
price_list = []
for i in range(2, MAX_PAGE_NUM+2):
    try:
        WebDriverWait(driver, some_timeout,ignored_exceptions=ignored_exceptions)\
                                .until(lambda browser: len(browser.find_elements_by_class_name("sc-wivooq-1")) >= 3)
        current_y = (driver.execute_script('return window.innerHeight') / 2) + driver.execute_script('return window.pageYOffset')
        while current_y < driver.execute_script('return document.body.scrollHeight') - 1000:
            driver.execute_script("window.scrollBy(0, 400)")
            current_y = (driver.execute_script('return window.innerHeight') / 2) + driver.execute_script('return window.pageYOffset')
            time.sleep(random.random()*0.8)
        
        my_xpath = "//a[@aria-label='Page " + str(i) + "']"
        # WebDriverWait(driver, some_timeout,ignored_exceptions=ignored_exceptions)\
        #                         .until(lambda browser: len(browser.find_elements_by_class_name("sc-wivooq-1")) >= int(min_len_per_page))
        
        elems = driver.find_elements_by_class_name("sc-wivooq-1")
        price_elems = driver.find_elements(by=By.CLASS_NAME, value="sc-hlnw2x-3")
        for elem in elems:
            name_list.append(elem.text)
        for elem in price_elems:
            price_list.append(elem.text)

        if len(elems) < 23:
            print("page " + str(i-1) + ": " + str(len(elems)))
            break
        for elem in elems:
            name_list.append(elem.text)
        print(i-1, len(elems))
        if i == MAX_PAGE_NUM+1:
            break
        button = driver.find_element_by_xpath("//a[@aria-label='Page " + str(i) + "']")
        button.click()

    except Exception as e:
        print(i-1)
        print(e)
```

```python
len(price_list)
names = name_list[:len(price_list)]
```

```python
data_ = {
    "Name": names,
    "Price": price_list
}
```

```python
pd.DataFrame(data=data_).to_csv("../data/properties_midland.csv")
```
