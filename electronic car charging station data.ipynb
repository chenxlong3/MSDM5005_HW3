```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://opendata.clp.com.hk/GetChargingSectionXML.aspx?lang=<LANG>'
response = get(url).text
html_soup = BeautifulSoup(response, 'html.parser')
```

```python
no = []
location = []
lat = []
lng = []
types = []
districtl = []
districts = []
address = []
provider = []
parkingno = []
for row in html_soup.find_all('station'):
    no.append(row.find('no').text)
    location.append(row.find('location').text)
    lat.append(row.find('lat').text)
    lng.append(row.find('lng').text)
    types.append(row.find('type').text)
    districtl.append(row.find('districtl').text)
    districts.append(row.find('districts').text)
    address.append(row.find('address').text)
    provider.append(row.find('provider').text)
    parkingno.append(row.find('parkingno').text)
```

```python
d = {'station No.':no,'location':location, 'latitude':lat, 'longtitude':lng,
       'type':types, 'districtL':districtl, 'districtS':districts, 
       'address':address, 'provider':provider, 'parking No.':parkingno} 
df = pd.DataFrame(data=d)
```
